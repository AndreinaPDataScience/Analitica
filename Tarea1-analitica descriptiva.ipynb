```python
#agregar una columna condicional a un dataframe
data1= data.assign(total_ventas =data["NA_Sales"]+data["EU_Sales"]+data["JP_Sales"]+data["Other_Sales"])
data1.head()
```

```python
condiciones= [
    (data1["Genre"]== "Simulation"),
    (data1["Genre"] == "Sports"),
    (data1["Genre"] == "Platforms"),
    (data1["Genre"] == "Raicing"),
    (data1["Genre"] == "Shooter"),
    (data1["Genre"] == "Misc"),
    (data1["Genre"] == "Action"),
    (data1["Genre"] == "Puzzle"),
    (data1["Genre"] == "Role-Playing")
]

valores_dados=[150,88,93,54,67,78,91,90,63]


data2= data1.assign(ganancias=np.select(condiciones,valores_dados,75))
```

```python
#alteracion al dataset, creando una version final

data=data2.assign(total_gan=data2["total_ventas"]*data2["ganancias"])
data.head()
```

```python
import numpy as np #Librería numérica
import matplotlib.pyplot as plt # Para crear gráficos con matplotlib
%matplotlib inline # Si quieres hacer estos gráficos dentro de un jupyter notebook
from sklearn.linear_model import LinearRegression #Regresión Lineal con scikit-learn
def f(x):  # función f(x) = 0.1*x + 1.25 + 0.2*Ruido_Gaussiano
    np.random.seed(42) # para poder reproducirlo
    y = 0.1*x + 1.25 + 0.2*np.random.randn(x.shape[0])
    return y
x2 = np.arange(0, 20, 0.5) # generamos valores x de 0 a 20 en intervalos de 0.5
y2 = f(x) # calculamos y a partir de la función que hemos generado
# hacemos un gráfico de los datos que hemos generado
plt.scatter(x2,y2,label='data', color='blue')
plt.title('Datos');
```

```python

from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import statsmodels.api as am
import  statsmodels.formula.api as smf
from sklearn import linear_model
```

```python
x=data2.iloc[:,:3].values
y= data2.iloc[:,: 2].values

"""nota:  Debido a ciertos errores con las variables cualitativas, estas se 
borraran para poder llevar a cabo la regresion lineal y las predicciones de 
manera efectiva"""
#scatter plot de las variables x y y
plt.scatter(data["Global_Sales"],data["EU_Sales"])
plt.xlabel('Ventas globales')
plt.ylabel('Ventas en Estados Unidos')
plt.show()
```

```python
#Separación de los datos en entrenamiento y prueba para los algoritmos
from sklearn.model_selection import train_test_split #libreria utilizada
#_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.19, random_state = 0)#,train_size=0.65, test_size =0.35,random_state=101)
#algoritmo que se va a utilizar

modelll= linear_model.LinearRegression()
#x=x.reshape(1, -1)
#est=x_test.reshape(1, -1)
#x_train=x_train.reshape(1, -1)
#y_train= y_train.reshape(1, -1)
#y_test= y_test.reshape(1, -1)
#entrenamiento del modelo
modelll.fit(x,y)
#prediccion
prediccion=modelll.predict(x)
```

```python
#error cuadratico medio
mse=mean_squared_error(y_true= y, y_pred=prediccion)
print(mse)

#la  raiz del mse
rmse= np.sqrt(mse)
#print de los resultados
print('El error cuadritico medio (MSE) = ' + str(mse))
print('Raíz  del error cuadratico medio (RMSE) = ' + str(rmse))

#r2 score
from sklearn.metrics import r2_score
r2= r2_score(prediccion, y)
```
